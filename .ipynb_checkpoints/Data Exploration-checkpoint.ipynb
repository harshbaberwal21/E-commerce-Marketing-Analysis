### Dimensionality Reductions
##### Running PCA to see if there is scope for reducing data dimensions

Note: For PCA, we will drop the aggregate features engineered earlier. If through PCA we are able to decrease the dimension significantly, we will move with this selection.

```python
# CampAccepted, Family_Size, Children
to_drop = ['ID','CampAccepted','Family_Size', 'Children']
temp = modeling_data.drop(to_drop, axis = 1)
```

```python
temp.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# Initializing and fitting  a min-max scaler to transform the data range
scaler = MinMaxScaler(feature_range=(-1, 1))
X = scaler.fit_transform(temp)


# Initializing PCA
pca = PCA()

# Fitting PCA to data and tranforming it
pca.fit(X)
transformed = pca.transform(X)
features = range(pca.n_components_)

# The below line should print 1 (or very very close to 1) 
# since this gives the variance explained by all the components

print(pca.explained_variance_ratio_.sum())

# Plotting Cumulative explained variance percentage against the number of components.
sns.set_style("whitegrid")
plt.plot(features, np.cumsum(pca.explained_variance_ratio_)*100)
plt.axhline(y=95, c = "g")
plt.xlabel("# of Components")
plt.ylabel("Cumulative Explained Variance Proportion")
plt.show()

print("# Components to retain >95% variance in the data: {}".format(sum(np.cumsum(pca.explained_variance_ratio_)<0.95)+1))
```

With 16 components, the transformed data is able to explain more than 95% of the variance, reducing the number of features from 24 to 16

### K-Means Clustering


##### Plotting the elbow curve with k-means cost (distortion) and silhouette score to identify the number of clusters

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score



def elbow_curve(data, n, k_max = 10):
    i_list = []
    s_list = []
    scaler = MinMaxScaler(feature_range=(-1, 1))
    X = scaler.fit_transform(data)
    
    pca = PCA(n_components=n)
    pca.fit(X)
    X_t = pca.transform(X)
    
    for i in range(2,k_max+1):
        model = KMeans(n_clusters=i, max_iter=1000)
        model.fit(X_t)
        i_list.append(model.inertia_)
        s_list.append(silhouette_score(X_t, model.labels_))
        
    ax1 = plt.subplot()
    ax2 = ax1.twinx()
    ax1.plot( [x for x in range(2,k_max+1)] , i_list, color = 'green')
    ax2.plot( [x for x in range(2,k_max+1)] , s_list, color = 'red')    
    plt.show()
    return i_list, s_list
```

```python
figure(figsize=(5, 3), dpi=150)
sns.set_style("white")
elbow_curve(data = temp, k_max = 10, n = 16)
```

**Since the model cost (distortion) decreases smoothly, we will use silhouette score to identify the optimum number of clusters in the sampled customer base.**

Looking at the graph, we can see that the silhouette score is maximum for 3 clusters and drops very low at 4 clusters. Hence, we will move forward with 3 clusters.

## Now given the accuracy of the model, the question comes how many ads should be run to be on a more profitable side compared to the intial investment made.

Assuming the cost and revenue grows linearly with the size of the target audience. Probably, another function or logic could be added here.

```python
cost = []
revenue = []
profit = []

for n in range(1,10001):
    
    cf_n = round_n(cf*n/n_test_customers)
    tp_n = cf_n[1][1]
    tn_n = cf_n[0][0]
    fp_n = cf_n[0][1]
    fn_n = cf_n[1][0]
    
    cost_n = (tp_n + fp_n)*cost_per_customer
    revenue_n = rev_per_accepted*tp_n
    profit_n = (rev_per_accepted*tp_n)- (tp_n + fp_n)*cost_per_customer
    
    cost.append(cost_n)
    revenue.append(revenue_n)
    profit.append(profit_n)






    
```
