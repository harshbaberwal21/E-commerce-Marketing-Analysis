##### Reading in the data

### Attributes

##### People
**ID**:  *Customer's unique identifier*  
**Year_Birth**:  *Customer's birth year*  
**Education**:  *Customer's education level*  
**Marital_Status**:  *Customer's marital status*  
**Income**:  *Customer's yearly household income*  
**Kidhome**:  *Number of children in customer's household*  
**Teenhome**:  *Number of teenagers in customer's household*  
**Dt_Customer**:  *Date of customer's enrollment with the company*  
**Recency**:  *Number of days since customer's last purchase*  
**Complain**:  *1 if the customer complained in the last 2 years, 0 otherwise*  
  
##### Products  
**MntWines**:  *Amount spent on wine in last 2 years*  
**MntFruits**:  *Amount spent on fruits in last 2 years*  
**MntMeatProducts**:  *Amount spent on meat in last 2 years*  
**MntFishProducts**:  *Amount spent on fish in last 2 years*  
**MntSweetProducts**:  *Amount spent on sweets in last 2 years*  
**MntGoldProds**:  *Amount spent on gold in last 2 years*  
  
##### Promotion  
**NumDealsPurchases**:  *Number of purchases made with a discount*  
**AcceptedCmp1**:  *1 if customer accepted the offer in the 1st campaign, 0 otherwise*  
**AcceptedCmp2**:  *1 if customer accepted the offer in the 2nd campaign, 0 otherwise*  
**AcceptedCmp3**:  *1 if customer accepted the offer in the 3rd campaign, 0 otherwise*  
**AcceptedCmp4**:  *1 if customer accepted the offer in the 4th campaign, 0 otherwise*  
**AcceptedCmp5**:  *1 if customer accepted the offer in the 5th campaign, 0 otherwise*  
**Response**:  *1 if customer accepted the offer in the last campaign, 0 otherwise*  
  
##### Place  
**NumWebPurchases**:  *Number of purchases made through the company’s website*  
**NumCatalogPurchases**:  *Number of purchases made using a catalogue*  
**NumStorePurchases**:  *Number of purchases made directly in stores*  
**NumWebVisitsMonth**:  *Number of visits to company’s website in the last month*

## Data Manipulation and Exploratory Data Analysis
##### (of sampled customers)

### Education Level

```python
cust_data['Education'].unique()
```

##### Coercing 2n Cycle with Master education level using map method of a series

```python
edu_map = {'Graduation':'Graduation', 'PhD':'PhD', 'Master':'Master', 'Basic':'Basic', '2n Cycle':'Master'}
cust_data['Education'] = cust_data['Education'].map(edu_map)
```

##### Plotting number of Customers vs Education Level

```python
edu_plot = sns.catplot(x = 'Education', data = cust_data, order =  ['Basic', 'Graduation', 'Master', 'PhD'], kind = 'count',  height=4, aspect=1.2)
edu_plot.set(xlabel='Education Level', ylabel = '# of Customers')
plt.show()
```

### Marital Status

##### Values that this attribute take and their respective frequencies

```python
cust_data['Marital_Status'].value_counts()
```

##### Recoding the Marital Status to few possible and sensible values.

```python
marital_map = {'Single':'Single', 'Together':'Partnered', 'Married':'Partnered', 'Divorced':'Single', 'Widow':'Single', 'Alone':'Single'}
cust_data['Marital_Status'] = cust_data['Marital_Status'].map(marital_map)
```

##### Dropping the rows with NA marital status ('absurd' and 'YOLO')

```python
cust_data = cust_data[~cust_data['Marital_Status'].isna()]
```

##### Plotting Number of Customers vs Marital Status

```python
mar_plot  = sns.catplot(x = 'Marital_Status', data = cust_data, kind = 'count')
mar_plot.set(xlabel='Marital Status', ylabel = '# of Customers', title = "# Customers by Marital Status")

print("Distribution of customers by marital Status")
print(cust_data['Marital_Status'].value_counts(normalize=True).apply(lambda x: round(x,3))*100,"\n")

plt.show()
```

### Age Distribution

##### Plotting historgram of age attribute

```python
sns.histplot(cust_data["Age"], kde = False)
plt.show()
```

##### Dopping Outlier ages in data. Keeping all below or equal to 80

```python
c = len(cust_data)
cust_data = cust_data.loc[cust_data["Age"] <= 80,:]
print("# Customers of or below the 80 years old: {}".format(len(cust_data)))
print("# Customers Dropped from Analysis: {}".format(c - len(cust_data)))
del(c)
```

##### Plotting histogram of age attribute post removing outliers

```python
sns.histplot(cust_data["Age"], kde = False)
plt.title("Age Distribution (Post Outlier Treatment)")
plt.show()
```

With the outliers removed, the sampled customers age variable is somewhat normal and seems like with a large sample it would be normal. It makes sense since most of the natural processes occur from a normal distribution.

### Campaign Analysis

##### Add legend (or xticks) on both rows

```python
camp_variables = ["AcceptedCmp1","AcceptedCmp2","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5","Response"]
```

```python
# Melting the dataset to long format for easy visulizing using seaborn's col and col_wrap arguments
a = cust_data.melt(id_vars=['ID'],value_vars=['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4','AcceptedCmp5', 'Response'],var_name=['Campaign'], value_name='Acceptance')
a["Campaign"] = a["Campaign"].map({'AcceptedCmp1':'1', 'AcceptedCmp2':'2', 'AcceptedCmp3':'3', 'AcceptedCmp4':'4','AcceptedCmp5':'5', 'Response':'Planned'})
sns.catplot(x="Acceptance", data=a, kind="count", col = 'Campaign', col_wrap = 3)
plt.subplots_adjust(hspace=0.3)
plt.xticks([0, 1], ['False','True'])
plt.show()
del(a)
```

The planned campaign performed really well in the samples customer base. Campaign 2 performed the worst but it could be because of the sampling as well. More data (or access to population data) would be highly desirable in such a condition.

### Marketplaces vs Purchases Analysis

```python
purchase_variables = ["NumWebPurchases","NumCatalogPurchases","NumStorePurchases","NumDealsPurchases"]
```

##### Plotting the distributions of number of purchases made by customers across the different marketplaces

```python
# Melting the dataset to long format for easy visulizing using seaborn's col and col_wrap arguments
a = cust_data.melt(id_vars=['ID'],value_vars=purchase_variables,var_name=['PurchaseType'], value_name='Purchases')
a["PurchaseType"] = a["PurchaseType"].map(dict(list(zip(purchase_variables, ["Web", "Catalog", "Store","Deals"]))))
sns.catplot(y="Purchases", data=a, kind="box", col = 'PurchaseType')
plt.subplots_adjust(hspace=0.3)
plt.show()
del(a)

```

On average, store and website purchases are made the most.

### Income vs Relationship Status X #Children at home

##### Plotting a crosstab between relationship status and number of children, depicting the average income per group

```python

relationship_vs_child_crosstab = cust_data.pivot_table(values="Income", index="Marital_Status", columns="Children", aggfunc=[np.mean]) ## - fill_value=0, margins=True
print(relationship_vs_child_crosstab)

figure(figsize=(8,3), dpi = 100)
h_map = sns.heatmap(pd.DataFrame(relationship_vs_child_crosstab.values,
                                 columns = relationship_vs_child_crosstab.columns.droplevel(),
                                 index = relationship_vs_child_crosstab.index,
                                ), 
                    annot = True, cmap="BuGn", fmt='g', annot_kws={"size":15, 'fontstyle':'italic'}, cbar = False)
h_map.set(ylabel = "Marital Status", xlabel = "# Children at Home", title = "Average Income")

```

Whether single or in relationship, customers with 0 children at home have the highest average income.

### Income vs Expenditure

```python
spend_variables = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds"]
spend_names = ["Wines","Fruits","Meat Products","Fish Products","Sweet Products","Gold Products"]
```

```python
temp = spend_variables.copy()
temp.extend(["Income", "Amount_Total"])
sns.heatmap(cust_data.loc[:,temp].corr(), annot = True, annot_kws={"size":10, "fontstyle":"italic"})
del(temp)
```
